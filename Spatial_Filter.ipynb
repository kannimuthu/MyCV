```python
#Mean Filter
import cv2
import numpy as np
import scipy.ndimage
import matplotlib.pyplot as plt
# Opening the image using cv2.
a = cv2.imread('C:/Users/Kanni/Desktop/KCE-ECE-CVP-2023/Demo/Figures/ultrasound_muscle.png')
# Converting the image to grayscale.
a = cv2.cvtColor(a, cv2.COLOR_BGR2GRAY)
# Initializing the filter of size 5 by 5.
# The filter is divided by 25 for normalization.
k = np.ones((5,5))/25
# performing convolution
b = scipy.ndimage.filters.convolve(a, k)
# Writing b to a file.
cv2.imwrite('C:/Users/Kanni/Desktop/KCE-ECE-CVP-2023/Demo/Figures/mean_output.png', b)
plt.imshow(b)
plt.show()
```

```python
#Vertical Sobel
import cv2
from skimage import filters

# Opening the image.
a = cv2.imread('C:/Users/Kanni/Desktop/KCE-ECE-CVP-2023/Demo/Figures/cir.png')
# Converting a to grayscale.
a = cv2.cvtColor(a, cv2.COLOR_BGR2GRAY)
# Performing vertical Sobel.
b = filters.sobel_v(a)
min_b, max_b = b.min(), b.max()
b = 255*(b-min_b)/(max_b-min_b)
# Saving b.
cv2.imwrite('C:/Users/Kanni/Desktop/KCE-ECE-CVP-2023/Demo/Figures/sobel_vcir.png', b)
```

```python
#prewitt
import cv2
from scipy import ndimage

# Opening the image.
a = cv2.imread('C:/Users/Kanni/Desktop/KCE-ECE-CVP-2023/Demo/Figures/cir.png')
# Converting a to grayscale .
a = cv2.cvtColor(a, cv2.COLOR_BGR2GRAY)
# Performing Prewitt filter.
b = ndimage.prewitt(a)
min_b, max_b = b.min(), b.max()
b = 255.0*(b-min_b)/(max_b-min_b)
# Saving b.
cv2.imwrite('C:/Users/Kanni/Desktop/KCE-ECE-CVP-2023/Demo/prewitt_cir.png', b)
```

```python
import cv2
import numpy as np

import numpy as np
from PIL import Image
from skimage.filters import frangi

img = cv2.imread('C:/Users/Kanni/Desktop/KCE-ECE-CVP-2023/Demo/Figures/angiogram1.png')
img1 = np.asarray(img)
img2 = frangi(img1, black_ridges=True)
img3 = 255*(img2-np.min(img2))/(np.max(img2)-np.min(img2))
cv2.imwrite('C:/Users/Kanni/Desktop/KCE-ECE-CVP-2023/Demo/Figures/frangi_output.png', img3)
```
